```python
column2 = 'How often do you exercise on average?'
for index in range(len(df[column2])):
    if df.at[index,column2] == 'Not at all':
        df.loc[index, column2] = 0
    elif df.at[index,column2] == 'Once a month':
        df.loc[index, column2] = 1
    elif df.at[index,column2] == 'A few times a month':
        df.loc[index, column2] = 2
    elif df.at[index,column2] == '1-2 times a week':
        df.loc[index, column2] = 3
    elif df.at[index,column2] == '3-5 times a week':
        df.loc[index, column2] = 4
    else:
        df.loc[index, column2] = 5

column3 = 'How often do you shower on average?'
for index in range(len(df[column3])):
    if df.at[index,column3] == 'Less than once a week+':
        df.loc[index, column3] = 0
    elif df.at[index,column3] == 'Once every 5-7 days':
        df.loc[index, column3] = 1
    elif df.at[index,column3] == 'Once every 3-4 days':
        df.loc[index, column3] = 2
    elif df.at[index,column3] == 'Once every 2 days':
        df.loc[index, column3] = 3
    elif df.at[index,column3] == 'Once a day':
        df.loc[index, column3] = 4
    else:
        df.loc[index, column3] = 5

column4 = 'How often do you brush your teeth on average?'
for index in range(len(df[column4])):
    if df.at[index,column4] == 'Less than once a week+':
        df.loc[index, column4] = 0
    elif df.at[index,column4] == 'Once every 5-7 days':
        df.loc[index, column4] = 1
    elif df.at[index,column4] == 'Once every 3-4 days':
        df.loc[index, column4] = 2
    elif df.at[index,column4] == 'Once every 2 days':
        df.loc[index, column4] = 3
    elif df.at[index,column4] == 'Once a day':
        df.loc[index, column4] = 4
    else:
        df.loc[index, column4] = 5

column5 = 'How long do you spend on skincare on average?'
for index in range(len(df[column5])):
    if df.at[index,column5] == '0 hours':
        df.loc[index, column5] = 0
    elif df.at[index,column5] == '0-10 minutes':
        df.loc[index, column5] = 1
    elif df.at[index,column5] == '10-20 minutes':
        df.loc[index, column5] = 2
    elif df.at[index,column5] == '20-30 minutes':
        df.loc[index, column5] = 3
    elif df.at[index,column5] == '30-60 minutes':
        df.loc[index, column5] = 4
    else:
        df.loc[index, column5] = 5


column7 = 'On a scale of 1-6, how much energy do you have on average?'
for index in range(len(df[column7])):
    if df.at[index,column7] == '<4':
        df.loc[index, column7] = 0
    elif df.at[index,column7] == '4-5':
        df.loc[index, column7] = 1
    elif df.at[index,column7] == '5-6':
        df.loc[index, column7] = 2
    elif df.at[index,column7] == '6-7':
        df.loc[index, column7] = 3
    elif df.at[index,column7] == '7-8':
        df.loc[index, column7] = 4
    else:
        df.loc[index, column7] = 5

print(df[column2])
print(df[column3])
print(df[column4])
print(df[column5])
print(df[column6])
print(df[column7])
```
